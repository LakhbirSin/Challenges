```python
# importation Librairies
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import zscore
```

```python
# implementaion tableau
erp = pd.read_csv('erp.xlsx - Sheet1.csv')
```

```python
erp
```

```python
# implementaion tableau
liaison = pd.read_csv('liaison.xlsx - Sheet1.csv')
liaison
```

```python
# implementaion tableau et renom du colonne
web = pd.read_csv('web.xlsx - Sheet1.csv')
web = web.rename(columns = {"sku" : "id_web"})
web
```

```python
#liaison tableau
ELI = erp.merge(liaison)
ELI
```

```python
#liaison tableau
WELI = web.merge(ELI)
WELI
```

```python
# le chiffre d'affaire par produit
WELI = WELI.assign(TotalCa = lambda x : x.total_sales * x.price)
WELI
```

```python
# prend seulement les products sans attachment
Filter_prod = WELI[WELI["post_type"] == "product"]
Filter_prod
```

```python
#Total Chiffre d'affaire
Filter_prod["TotalCa"].sum()
```

```python
# suppression des doublons
WELI.drop_duplicates()
```

```python
# calcul zscores price
zscores = stats.zscore(WELI["price"])
zscores
```

```python
# Interquartiles
Q1,Q3 = np.percentile(WELI["price"],[25,75])
Q1,Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
# Ajout du colonne zscores
WELI = WELI.assign(zscores=zscores)
WELI
```

```python
WELI = WELI[pd.isnull(WELI["id_web"]) == False]
```

```python
Filter_prod = WELI[WELI["post_type"] == "product"]
Filter_prod.reset_index(inplace = True, drop = True)
Filter_prod
```

```python
# zcores supérieurs et égale a 2 seulement
Filter_Z = Filter_prod[Filter_prod["zscores"] >= 2]
Filter_Z
```

```python
# Nuage de points avec les outliers en rouge
import matplotlib.pyplot as plt
x = Filter_prod.index.to_list()
y = Filter_prod["price"]
a = Filter_Z.index.to_list()
b = Filter_Z["price"]
plt.scatter(x,y, color ='blue', marker='+')
plt.scatter(a,b, color ='red', marker='+')
```
